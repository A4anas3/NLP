## Name Entity Recognition

Named Entity Recognition

1 sentence "The Eiffel Tower was built from 1887 to 1889 by French engineer Gustave Eiffel, whose company specialized in building metal frameworks and structures.

Person Eg: Krish C Naik

Place Or Location Eg: India

5 Date Eg: September, 24-09-1989

6 Time Eg: 4:30pm

7 Money Eg: 1 million dollar

Organization Eg: INeuron Private Limited

Percent Eg: 20%, twenty percent



```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
from nltk.tokenize import word_tokenize
```

```python
word=word_tokenize(sentence)
```

```python
import nltk
tag_elements=nltk.pos_tag(word)
```

```python
nltk.download('maxent_ne_chunker_tab', quiet=True)
nltk.download('words', quiet=True)
```

```python

nltk.ne_chunk(tag_elements).draw()

```
