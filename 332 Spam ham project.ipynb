```python
import pandas as pd 
messages=pd.read_csv('SpamClassifier-master\smsspamcollection\SMSSpamCollection',sep='\t',names=["label","message"])
```

```python
messages.head()
```

```python
import re 
import nltk
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
```

```python
pr=PorterStemmer()
```

```python
reviews=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[pr.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    reviews.append(review)
```

```python
reviews
```

## Create a Bag of word


```python
y=pd.get_dummies(messages['label']).astype(int)
y=y.iloc[:,0].values
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(reviews,y,test_size=0.2,random_state=0)

```

```python
## Create a Bag of word
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=1000,formatter=dict(float=lambda x: "%.3f" % x))
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam =MultinomialNB()
```

```python
spam.fit(X_train,y_train)
```

```python
y_pred=spam.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
