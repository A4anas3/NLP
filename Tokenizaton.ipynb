```python
corpus="""Hello s'Welcome,to anas NLP world. 
NLP is fun.
NLP is interesting."""
```

```python
print(corpus)
```

```python

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize
sent_tokenizes=sent_tokenize(corpus)
```

```python
print(sent_tokenizes)
```

```python
for list in sent_tokenizes:
    print(list)
```

```python
from nltk.tokenize import word_tokenize
word_tokenizes=word_tokenize(corpus)
```

```python
print(word_tokenizes)
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenizes=wordpunct_tokenize(corpus)
print(wordpunct_tokenizes)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
Treee=TreebankWordTokenizer()
```

```python
Treee.tokenize(corpus)
```
